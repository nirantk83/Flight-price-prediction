```python
#To predict price of flight we have to convert all object to int,float to build and apply model on ths dataset
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
#we import the lable encoder from sklearn libret to convert object to int,float
```

```python
df['airline']=lr.fit_transform(df['airline'])
df['flight']=lr.fit_transform(df['flight'])
df['source_city']=lr.fit_transform(df['source_city'])
df['departure_time']=lr.fit_transform(df['departure_time'])
df['stops']=lr.fit_transform(df['stops'])
df['arrival_time']=lr.fit_transform(df['arrival_time'])
df['destination_city']=lr.fit_transform(df['destination_city'])
df['class']=lr.fit_transform(df['class'])
df['duration']=lr.fit_transform(df['duration'])
df['days_left']=lr.fit_transform(df['days_left'])
```

```python
df.info()
```

```python
df.drop(['flight','arrival_time'],axis=1,inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df['price']
#split datadet into x and Y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#model building
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
Lr.fit(xtrain,ytrain)
ypred=Lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```
